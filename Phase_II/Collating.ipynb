# Combining Input Events CV and MV

```python

cv = pd.read_csv("INPUTEVENTS_CV.csv")
mv = pd.read_csv("INPUTEVENTS_MV.csv")

cv = cv.drop([mv.columns[0]], axis=1)
cv = cv.drop(['cgid'], axis =1 )                        # removed the Caregiver variable (not available in the MV table)
mv = mv.drop([mv.columns[0]], axis=1)

print(mv.shape, cv.shape)
display(mv.head(), cv.head())
```

```python
tables = [mv, cv]
df = pd.concat(tables)
print(df.shape)
df.head()
```

```python
df.to_csv('/Users/maxb/Library/CloudStorage/OneDrive-UniversityofWaterloo/Hospital Research/Datasets/Cleaned MIMIC-III Dataset/INPUTEVENTS.csv')
os.remove("INPUTEVENTS_CV.csv")
os.remove("INPUTEVENTS_MV.csv")
```

# Input tbales have been combined

# Combining Output Events and Chart Events

```python
Charts = pd.read_csv('CHARTEVENTS.csv')
Outputs = pd.read_csv('OUTPUTEVENTS.csv')

Charts = Charts.drop([Charts.columns[0]], axis=1)
Outputs = Outputs.drop([Outputs.columns[0]], axis=1)

print(Charts.shape, Outputs.shape)
display(Charts.head(), Outputs.head())
```
